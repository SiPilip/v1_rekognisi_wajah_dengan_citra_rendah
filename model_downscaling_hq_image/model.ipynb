### Menurunkan kualitas gambar (HQ -> LQ)


```python
import torch
```

```python
from PIL import Image

image_path = "./../raw_hq_data/pilip_hq_1.jpg"
img = Image.open(image_path)
```

```python
from torchvision import transforms

IMG_SIZE = 50

downscaling_transforms = transforms.Compose([
  transforms.Resize((IMG_SIZE, IMG_SIZE), antialias=True),
  transforms.ToTensor()
])
```

```python
downscaling_result = downscaling_transforms(img)
```

```python
import matplotlib.pyplot as plt

plt.imshow(downscaling_result.permute(1,2,0))
plt.title("contoh gambar uy")
plt.axis(False)
```

```python
to_pil_img = transforms.ToPILImage()

pil_img = to_pil_img(downscaling_result)

pil_img.save("./../raw_lq_data/downscaled_img.jpg")
pil_img.show()
```
